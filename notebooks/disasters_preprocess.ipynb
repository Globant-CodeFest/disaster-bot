```python
import pandas as pd

disasters = pd.read_csv("../data/disaster_chile_located.csv")
disasters.head()
```

```python
disasters_chile = disasters[(disasters.Country == 'Chile') & (disasters.Latitude.notnull()) & (disasters.Longitude.notnull())]
disasters_chile.shape
```

```python
def parse_lat_long(coord_string):
    components = coord_string.split()
    num = float(components[0])
    return num
```

```python
# Apply this function to your latitude and longitude columns
disasters_chile['Latitude'] = disasters_chile['Latitude'].apply(parse_lat_long)
disasters_chile['Longitude'] = disasters_chile['Longitude'].apply(parse_lat_long)
```

```python
import geopandas as gpd
from shapely.geometry import Point
# Create a new geometry column in your DataFrame
disasters_chile['Coordinates'] = list(zip(disasters_chile.Longitude, disasters_chile.Latitude))
disasters_chile['Coordinates'] = disasters_chile['Coordinates'].apply(Point)

gdf = gpd.GeoDataFrame(disasters_chile, geometry='Coordinates')
```

```python
gdf.to_file('../data/distasters_parsed.geojson', driver='GeoJSON')
```
